```python
import random
import cv2
from matplotlib import pyplot as plt
```

```python
import cv2
img = cv2.imread("hermoine.jpg")
image_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Original")
plt.imshow(image_rgb)
```

Adding noise creates a salt and pepper sprinkled effect on the image

```python
def add_noise(img):
  row, col= img.shape
  number_of_pixels=random.randint(300,10000)
  for i in range(number_of_pixels):
    ycord=random.randint(0,row-1)
    xcord=random.randint(0,col-1)
    img[ycord][xcord]= 255
  number_of_pixels=random.randint(300,10000)
  for i in range(number_of_pixels):
    ycord=random.randint(0,row-1)
    xcord=random.randint(0,col-1)
    img[ycord][xcord]= 0

  return img
```

```python
image= cv2.imread('hermoine.jpg')
plt.imshow(image)
plt.show()
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show
```

Code to create salt and pepper effect on the image

```python
img= cv2.imread('hermoine.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imwrite('hermoinenoise.jpg',add_noise(img))
noise_image=cv2.imread('hermoinenoise.jpg')
plt.imshow(noise_image)
plt.show()
```

```python
median= cv2.medianBlur(noise_image,5)
plt.imshow(median)
plt.show()
```
