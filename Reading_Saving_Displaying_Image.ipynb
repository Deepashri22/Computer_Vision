```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
image = cv2.imread("einstein.jpg")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.title("Original")
plt.imshow(image_rgb)
```

SAVING THE GRAYSCALE IMAGE


```python
gray_image=cv2.imread("einstein.jpg",cv2.COLOR_BGR2GRAY)
cv2.imwrite("einstein_gray.jpg",gray_image)
plt.title("Gray Scale")
plt.imshow(gray_image)
```

LOADING AND DISLAYING THE SAVED IMAGE

```python
img=cv2.imread("einstein_gray.jpg")

plt.title("Saved Image")
plt.imshow(img)
```
